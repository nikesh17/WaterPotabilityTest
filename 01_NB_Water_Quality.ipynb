```python
from scipy import stats
```

```python
water_data['pH_Z_Score'] = stats.zscore(water_data['pH'].dropna())
water_data['Iron_Z_Score'] = stats.zscore(water_data['Iron'].dropna())

pH_outliers = water_data[(water_data['pH_Z_Score'] > 3) | (water_data['pH_Z_Score'] < -3)]
Iron_outliers = water_data[(water_data['Iron_Z_Score'] > 3) | (water_data['Iron_Z_Score'] < -3)]

print(f'Number of pH outliers: {pH_outliers.shape[0]}')
print(f'Number of Iron outliers: {Iron_outliers.shape[0]}')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
scaler = LabelEncoder()
```

```python
for column in enc:
    water_data[column+'ENC'] = scaler.fit_transform(water_data[column])
```

```python
water_data
```

```python
water_data = water_data.select_dtypes(exclude='object')
```

```python
water_data.info()
```

```python
water_data.corr()['Target']
```

```python
water_data.isna().sum()
```

```python
water_data_mean = water_data.drop(columns = ['Day', 'Time of Day', 'Target','ColorENC', 'SourceENC','MonthENC'])
```

```python
for column in water_data_mean:
    mean_value = water_data[column].mean()
    water_data[column].fillna(mean_value,inplace= True)
```

```python
water_data.isna().sum()
```

```python
water_data_mode = ['Day','Time of Day']
```

```python
for column in water_data_mode:
    mode_value = water_data[column].mode()[0]
    water_data[column].fillna(mode_value, inplace= True)
```

```python
water_data.isna().sum()
```

```python
water_data.corr()['Target']
```

```python
y = water_data['Target']
x = water_data.drop(['Target','Index'], axis = 1)
```

```python
x
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier()
```

```python
X_train,X_test,y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42, stratify = y)
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
model_accuracy = accuracy_score(y_test, predictions)
model_accuracy
```

```python
gradient_model = GradientBoostingClassifier()
```

```python
gbmodel = gradient_model.fit(X_train, y_train)
gbmodel
```

```python
g_predictions = gbmodel.predict(X_test)
g_predictions
```

```python
accuracy_score(y_test, g_predictions)
```
